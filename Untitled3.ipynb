```python
#loading the necessary datasets
leg_dataset=pd.read_csv("leg_dataset.csv")
phi_dataset=pd.read_csv("phi_dataset.csv")
```

```python
leg_dataset.head()
```

```python
phi_dataset.head()
```

```python
main_dataset=pd.concat([leg_dataset,phi_dataset]).reset_index(drop=True)
main_dataset.head()
```

```python
y=main_dataset['Label'];
X=main_dataset.drop(['Label'],axis=1)
X.shape
```

```python
y.shape
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=60)
X_train.shape
```

```python
X_test.shape
```

```python
xgb=XGBClassifier(learning_rate=0.4,max_depth=4)
xgb.fit(X_train,y_train)
```

```python
y_train_xgb=xgb.predict(X_train)
y_test_xgb=xgb.predict(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=19)
```

```python
knn.fit(X_train,y_train)
y_train_knn=knn.predict(X_train)
y_test_knn=knn.predict(X_test)
```

```python
train_acc_knn=accuracy_score(y_train_knn,y_train)
test_acc_knn=accuracy_score(y_test,y_test_knn)
train_acc_knn
```
