```python
import pandas as pd
```

```python
leg_dataset=pd.read_csv("Datasets/leg_dataset11.csv")
leg_dataset
```

```python
for i in range(5000):
    if(leg_dataset.loc[i,'URL_Length']<54):
        leg_dataset.loc[i,'URL_Length']=0
    else:
        leg_dataset.loc[i,'URL_Length']=1
    if(leg_dataset.loc[i,'URL_Depth']>5):
        leg_dataset.loc[i,'URL_Depth']=1
    else:
        leg_dataset.loc[i,'URL_Depth']=0
    if(leg_dataset.loc[i,'Web_Traffic']<100000):
        leg_dataset.loc[i,'Web_Traffic']=0
    else:
        leg_dataset.loc[i,'Web_Traffic']=1
    ageofdomain=leg_dataset.loc[i,'Domain_Age']
    if ((ageofdomain/30) < 6):
      leg_dataset.loc[i,'Domain_Age']=1
      leg_dataset.loc[i,'Domain_End']=1
    else:
      leg_dataset.loc[i,'Domain_End']=0
      leg_dataset.loc[i,'Domain_Age']=0
    if(leg_dataset.loc[i,'Redirection']>6 and leg_dataset.loc[i,'Redirection']>7):
        leg_dataset.loc[i,'Redirection']=1
    else:
        leg_dataset.loc[i,'Redirection']=0
```
